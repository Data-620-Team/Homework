```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

G = nx.Graph()
```

```python
g = nx.generators.small.krackhardt_kite_graph()
```

```python
g.number_of_edges()
```

```python
print('Nodes from krackhardt_kite_graph: ', '\n', g.nodes(), '\n')

print('Edges frp, krackhardt_kite_graph: ', '\n', g.edges())
```

```python
g.number_of_nodes()
```

```python
nx.draw(g, with_labels=1)
plt.show()
```

```python
G1 = nx.Graph()

e1 = g.edges()
G1.add_edges_from(e1)

print("Edges for Name network:", "\n", G1.edges())
```

```python
H=nx.relabel_nodes(G1, person)
```
